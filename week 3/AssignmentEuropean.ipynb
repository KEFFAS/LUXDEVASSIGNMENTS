### Assignment Solution: European Countries not in EU (Using Sets & Pandas)
#### 1.Problem Summary
#### Task: Determine which European countries (in dataset) are not EU members
#### Tools: pandas, Python sets


```python
## Importing necessary libraries
import pandas as pd
europe=pd.read_excel('Europe_and_EU.xlsx',sheet_name='Europe')
eu=pd.read_excel('Europe_and_EU.xlsx',sheet_name='EU')
```

```python
# Clean the Country column by removing leading/trailing whitespaces
europe['Country'] = europe['Country'].str.strip()
europe['Country'] = europe['Country'].str.strip()
```

```python
# Top five rows (europe)
europe.head()
```

```python
# Top five rows (eu)
eu.head()
```

```python
# Convert to sets for comparison
europe_countries = set(europe['Country'].dropna().unique())
eu_countries = set(eu['Country'].dropna().unique())
```

```python
# Check if EU countries are a subset of European countries
is_subset = eu_countries.issubset(europe_countries)
print(is_subset)
```

```python
# Find European countries NOT in EU
non_eu_countries = europe_countries - eu_countries
print("European countries NOT in EU (2018):")
for country in sorted(non_eu_countries): #loops through each element in the collection.
    print(country)
```
